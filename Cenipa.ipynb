# Pacotes Necessários

```python
import pandas as pd
```

# Iniciando o ETL (Extract)

## Lendo os Dados

```python
dataframe = pd.read_csv("http://sistema.cenipa.aer.mil.br/cenipa/media/opendata/ocorrencia.csv", 
                        sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
dataframe.head()
dataframe.tail()
```

## Removendo as colunas desnecessárias

```python
df = dataframe.drop(['codigo_ocorrencia1', 'codigo_ocorrencia2', 'codigo_ocorrencia3', 'codigo_ocorrencia4'], axis=1)
#axis = 1 indica que removeremos as colunas
df.head()
```

## Visualizando os tipos de dados

```python
df.dtypes
df.ocorrencia_dia #Visualizando apenas uma coluna
df.ocorrencia_dia.dt.month
```

## Visualizando uma amostra randômica dos dados

```python
df.sample(n=10) #Carregue n registros de forma randômica
df.sample(frac=0.005) #Retorna uma fração percentual da base
```
